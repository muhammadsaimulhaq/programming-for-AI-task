### Implement the N-Queens Problem (Dynamic)

```python
def is_safe(board, row, col, n):
    # Check this column on upper side
    for i in range(row):
        if board[i][col] == 1:
            return False
    
    # Check upper diagonal on left side
    i, j = row, col
    while i >= 0 and j >= 0:
        if board[i][j] == 1:
            return False
        i -= 1
        j -= 1
    
    # Check upper diagonal on right side
    i, j = row, col
    while i >= 0 and j < n:
        if board[i][j] == 1:
            return False
        i -= 1
        j += 1
    
    return True


def print_board(board, n):
    print("\nSolution:")
    for i in range(n):
        for j in range(n):
            if board[i][j] == 1:
                print(" Q ", end="")
            else:
                print(" . ", end="")
        print()

def solve_n_queens_util(board, row, n, solutions):
    # Base case: If all queens are placed
    if row == n:
        solutions.append([row[:] for row in board])  # save a copy
        return
    
    # Try placing queen in every column of this row
    for col in range(n):
        if is_safe(board, row, col, n):
            # Place the queen
            board[row][col] = 1
            
            # Recur to place rest of the queens
            solve_n_queens_util(board, row + 1, n, solutions)
            
            # If we didn't get solution → backtrack
            board[row][col] = 0


def solve_n_queens(n):
    # Create empty chessboard (n × n)
    board = [[0 for _ in range(n)] for _ in range(n)]
    solutions = []
    
    print(f"\nSolving {n}-Queens Problem...")
    solve_n_queens_util(board, 0, n, solutions)
    
    if not solutions:
        print("No solution exists!")
    else:
        print(f"\nFound {len(solutions)} solution(s):")
        for idx, sol in enumerate(solutions, 1):
            print(f"\nSolution {idx}:")
            print_board(sol, n)

if __name__ == "__main__":
    try:
        n = int(input("Enter number of queens (board size): "))
        if n < 1:
            print("Please enter a positive number!")
        elif n == 2 or n == 3:
            print(f"No solution exists for {n}-Queens.")
        else:
            solve_n_queens(n)
    except ValueError:
        print("Please enter a valid integer!")
```
