### Implement the Water Jug Problem using Depth-First Search
(DFS) and print the rules applied at each step.

```python
# Water Jug Problem using DFS
# Goal: Get exactly 2 liters in one of the jugs
# Jugs: 4L and 3L

def print_state(state, action=""):
    x, y = state
    if action:
        print(f"→ {action:25}  →  ({x}, {y})")
    else:
        print(f"Initial state:             ({x}, {y})")

def is_goal(state, target=2):
    x, y = state
    return x == target or y == target

def get_successors(state, cap1=4, cap2=3):
    x, y = state
    moves = []
    
    # 1. Fill jug 1 completely
    moves.append( ((cap1, y), f"Fill Jug1 ({cap1}L)") )
    
    # 2. Fill jug 2 completely
    moves.append( ((x, cap2), f"Fill Jug2 ({cap2}L)") )
    
    # 3. Empty jug 1
    moves.append( ((0, y), "Empty Jug1") )
    
    # 4. Empty jug 2
    moves.append( ((x, 0), "Empty Jug2") )
    
    # 5. Pour from jug1 → jug2
    pour_to_2 = min(x, cap2 - y)
    moves.append( ((x - pour_to_2, y + pour_to_2), "Pour Jug1 → Jug2") )
    
    # 6. Pour from jug2 → jug1
    pour_to_1 = min(y, cap1 - x)
    moves.append( ((x + pour_to_1, y - pour_to_1), "Pour Jug2 → Jug1") )
    
    return moves

def dfs_water_jug(start=(0,0), target=2, cap1=4, cap2=3):
    stack = [(start, [])]           # (current_state, path_of_actions)
    visited = set()
    
    print("\nWater Jug Problem (DFS) - 4L and 3L jugs, target =", target, "L")
    print_state(start)
    
    while stack:
        current, path = stack.pop()   # pop from end → DFS
        
        if current in visited:
            continue
            
        visited.add(current)
        
        # Show the state after last action
        if path:
            last_action = path[-1]
            print_state(current, last_action)
        
        if is_goal(current, target):

            print("SOLUTION FOUND!")
            print("Final state:", current)
            print("Steps taken:", len(path))
            print("Sequence of actions:")
            for i, action in enumerate(path, 1):
                print(f"{i:2d}. {action}")
            return True
        
        # Generate next states
        for next_state, action in get_successors(current, cap1, cap2):
            if next_state not in visited:
                new_path = path + [action]
                stack.append((next_state, new_path))
    
    print("\nNo solution found.")
    return False


# Run the solver
if __name__ == "__main__":
    dfs_water_jug(start=(0,0), target=2)
```
